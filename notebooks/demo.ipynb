# VisFEM

This notebook launches the **VisFEM** web application inside Jupyter using [Trame](https://kitware.github.io/trame/).

VisFEM is a browser-based viewer for finite-element meshes and simulation results.  
It is built on [PyVista](https://docs.pyvista.org/) + [VTK](https://vtk.org/) for rendering and [Trame](https://kitware.github.io/trame/) + [Vuetify 3](https://vuetifyjs.com/) for the UI.

> **Requirement:** run Jupyter from the repo root so that the `data/` directory is on the Python path.

## 1. Environment

```python
import importlib.metadata

import pyvista as pv

from visfem.app import VisfemApp

print(f"PyVista : {pv.__version__}")
print(f"Trame   : {importlib.metadata.version('trame')}")
```

## 2. Start the VisFEM server

```python
import asyncio

print("Importing visfem...")
print("Import done. Initializing app...")

app = VisfemApp()
print("App ready. Starting server...")

app.server.start(exec_mode="task", open_browser=False, port=0)
await asyncio.sleep(2) # time buffer to bind the port

print(f"VisFEM server running on port {app.server.port}")
```

## 3. Embed the webapp

```python
from IPython.display import IFrame, display

url = f"http://localhost:{app.server.port}"
print(f"Opening: {url}")
display(IFrame(src=url, width="100%", height="720px"))
```
